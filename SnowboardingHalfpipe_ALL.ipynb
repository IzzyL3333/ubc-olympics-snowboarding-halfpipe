# Snowboarding Halfpipe ALL 2002-2014 Score

```python
import pandas as pd # library for data manipulation and analysis
import numpy as np # library for arrays of data
import matplotlib.pyplot as plt # library for plotting data 
from sklearn.linear_model import LinearRegression # library for Linear Regression calculations and analysis
import statsmodels.api as sm # library for Linear Regression calculations and analysis 
```

### Raw Data for Snowboarding Halfpipe Athletes Adjusted Score vs BMI from 2002 - 2014

```python
# Load Olympic Snowboarding Halfpipe Athletes Data
AthleteData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_ALL.csv") 
# X_BMI: BMI of Snowboarding Halfpipe athletes, Y_Score: Adjusted Score of Snowboarding Halfpipe athletes 
X_BMI = AthleteData . iloc[:, 17 ] . to_numpy ()
Y_Score = AthleteData . iloc [:, 16 ] . to_numpy ()
# plot the raw data of Snowboarding Halfpipe Athletes Adjusted Score vs BMI 
plt . figure ( figsize = (10 , 6) )
plt . scatter ( X_BMI, Y_Score, c='blue')
plt . xlabel ( 'BMI (kg/m^2)' ) 
plt . ylabel ( 'Score Adjusted' ) 
plt . title ( 'Top 12 Snowboarding Halfpipe Athletes Adjusted Score vs BMI (2002-2014)') 
plt . show ()
```

### Inferential Analysis for Snowboarding Halfpipe Athletes Score vs BMI from 2002 - 2014

```python
# Load Olympic Snowboarding Halfpipe Athletes Data
AthleteData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_ALL.csv") 
# X_BMI: BMI of Snowboarding Halfpipe athletes, Y_Score: Adjusted Score of Snowboarding Halfpipe athletes 
X_BMI = AthleteData . iloc[:, 17 ] . to_numpy ()
Y_Score = AthleteData . iloc [:, 16 ] . to_numpy ()

# Linear regression on the BMI using the statsmodel 
# add a column of ones to the x variables (required for statsmodel regression)
X_BMI_new = sm.add_constant(X_BMI)

# fit linear model to training data (drop missing values)
AD_model = sm.OLS(Y_Score, X_BMI_new, missing='drop').fit()

# create variables for the trend, 95% confidence intervals, and Rsquare respectively
print('FINAL OUTCOMES:')
AD_trend = round(AD_model.params[1],3)
AD_conf = AD_model.conf_int()[1]
AD_r2 = round(AD_model.rsquared,4)
AD_Yint = round(AD_model.params[0])
print(" - Adjusted Score TREND: " + str(AD_trend))
print(" - Adjusted Score 95% CONFIDENCE INTERVAL: " + str(AD_conf))
print(" - Adjusted Score R^2: " + str(AD_r2))
AD_equation = "Y = " + str(AD_trend) + ' /(kg/m^2) + ' + str(AD_Yint)
print(AD_equation)

# Plotting the data and linear regression
plt . figure ( figsize = (10 , 6) )
plt . scatter ( X_BMI , Y_Score, label = 'Adjusted Score vs BMI', c='blue')
plt.plot(X_BMI, AD_model.fittedvalues,color ='orange', label = 'Linear Regression Model')
# annotate the plot with the trend and R^2
plt.annotate('R-squared: ' + str(AD_r2) + "\n", xy=(0.7, 0.1), xycoords='axes fraction', color='m')
plt . xlabel ( 'BMI (kg/m^2)' ) 
plt . ylabel ( 'Adjusted Score' ) 
plt . title ( 'Top 12 Snowboarding Halfpipe Athletes Adjusted Score vs BMI (2002-2014)' ) 
plt . legend()
plt . show ()
```

### Box Plot for Top 12 Olympic Snowboarding Halfpipe Athletes Adjusted Score vs Gender

```python
# Load Olympic Snowboarding Halfpipe Athletes Data
MaleData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_Men.csv") 
FemaleData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_Women.csv")

# Male_Score: Adjusted Score of Top 12 Olympic Snowboarding Halfpipe Male Athletes 2002-2014
Male_Score = MaleData. iloc [:, 16 ] . to_numpy ()
# Female_Score: Adjusted Score of Top 12 Olympic Snowboarding Halfpipe Female Athletes 2002-2014
Female_Score =  FemaleData. iloc [:, 16 ] . to_numpy ()

# Calculate Min, Max, Average for Top 12 Male and Female Adjusted Score
print('2002-2014 TOP 12 MALE ADJUSTED SCORE RESULTS:')
print(' - MIN: ' + str(Male_Score.min(0))) 
print(' - MAX: ' + str(Male_Score.max(0)))
print(' - AVG: ' + str(round(Male_Score.mean(0),1)))
print(' - STD: ' + str(round(Male_Score.std(0),1)))
print('2002-2014 TOP 12 FEMALE ADJUSTED SCORE RESULTS:')
print(' - MIN: ' + str(Female_Score.min(0))) 
print(' - MAX: ' + str(Female_Score.max(0)))
print(' - AVG: ' + str(round(Female_Score.mean(0),1)))
print(' - STD: ' + str(round(Female_Score.std(0),1)))

# Creating boxplot with Male_Score and Women_Score
fig, ax1 = plt.subplots(figsize=(10, 6))
plt.boxplot([Male_Score,Female_Score])
plt.xticks([1, 2], ['Male','Female'])
ax1.set(
    axisbelow=True,  # Hide the grid behind plot objects
    title='Adjusted Score of Top 12 Olympic Snowboarding Halfpipe Male Athletes vs Female Athletes (2002-2014)',
    xlabel='Gender',
    ylabel='Adjusted Score'
)
plt.show()
```
