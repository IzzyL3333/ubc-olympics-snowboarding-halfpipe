# Snowboarding Halfpipe Men 2002-2014 Score

```python
import pandas as pd # library for data manipulation and analysis
import numpy as np # library for arrays of data
import matplotlib.pyplot as plt # library for plotting data 
from sklearn.linear_model import LinearRegression # library for Linear Regression calculations and analysis
import statsmodels.api as sm # library for Linear Regression calculations and analysis 
```

### Raw Data for Top 12 Snowboarding Halfpipe Men Athletes Adjusted Score vs BMI from 2002 - 2014

```python
# Load Olympic Top 12 Snowboarding Halfpipe Men Athletes Data
MenData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_Men.csv") 
# X_BMI: BMI of Men Athlete, Y_Score: Adjusted Score of Men Athlete
X_BMI = MenData . iloc[:, 17 ] . to_numpy ()
Y_Score = MenData . iloc [:, 16 ] . to_numpy ()
# plot the raw data of Men Adjusted Score vs BMI 
plt . figure ( figsize = (10 , 6) )
plt . scatter ( X_BMI, Y_Score, c='blue')
plt . xlabel ( 'BMI (kg/m^2)' ) 
plt . ylabel ( 'Score Adjusted' ) 
plt . title ( 'Top 12 Snowboarding Halfpipe Male Athletes Adjusted Score vs BMI (2002-2014)') 
plt . show ()
```

### Inferential Analysis for Top 12 Snowboarding Halfpipe Men Athletes Score vs BMI from 2002 - 2014

```python
# Load Olympic Top 12 Snowboarding Halfpipe Men Athletes Data
MenData = pd . read_csv ( "SnowboardingHalfpipe_2002_2014_Men.csv") 
# X_BMI: BMI of Men Athlete, Y_Score: Adjusted Score of Men Athlete
X_BMI = MenData . iloc[:, 17 ] . to_numpy ()
Y_Score = MenData . iloc [:, 16 ] . to_numpy ()

# Linear regression on the BMI using the statsmodel 
# add a column of ones to the x variables (required for statsmodel regression)
X_BMI_new = sm.add_constant(X_BMI)

# fit linear model to training data (drop missing values)
MD_model = sm.OLS(Y_Score, X_BMI_new, missing='drop').fit()

# create variables for the trend, 95% confidence intervals, and Rsquare respectively
print('FINAL OUTCOMES:')
MD_trend = round(MD_model.params[1],3)
MD_conf = MD_model.conf_int()[1]
MD_r2 = round(MD_model.rsquared,4)
MD_Yint = round(MD_model.params[0])
print(" - Adjusted Score TREND: " + str(MD_trend))
print(" - Adjusted Score 95% CONFIDENCE INTERVAL: " + str(MD_conf))
print(" - Adjusted Score R^2: " + str(MD_r2))
MD_equation = "Y = " + str(MD_trend) + ' /(kg/m^2) + ' + str(MD_Yint)
print(MD_equation)

# Plotting the data and linear regression
plt . figure ( figsize = (10 , 6) )
plt . scatter ( X_BMI , Y_Score, label = 'Adjusted Score vs BMI', c='blue')
plt.plot(X_BMI, MD_model.fittedvalues,color ='orange', label = 'Linear Regression Model')
# annotate the plot with the trend and R^2
plt.annotate('R-squared: ' + str(MD_r2) + "\n", xy=(0.71, 0.1), xycoords='axes fraction', color='m')
plt . xlabel ( 'BMI (kg/m^2)' ) 
plt . ylabel ( 'Adjusted Score' ) 
plt . title ( 'Top 12 Snowboarding Halfpipe Male Athletes Adjusted Score vs BMI (2002-2014)' ) 
plt . legend()
plt . show ()
```
